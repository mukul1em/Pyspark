```
from pyspark import SparkConf, SparkContext
conf = SparkConf().setAppName('minmax')
sc = SparkContext.getOrCreate(conf=conf)
```

```
rdd = sc.textFile('/FileStore/tables/movie_review.csv')
rdd.collect()
```

```
#converting it into a key value pair
def keyvaluepair(sent):
    name, rate = sent.split(',')
    return (name, int(rate))

rdd2 = rdd.map(keyvaluepair)
rdd2.collect()
```

```
rdd2.reduceByKey(lambda x,y: min(x,y)).collect()
```

```
rdd2.reduceByKey(lambda x,y: max(x,y)).collect()
```

```
rdd2.groupByKey().mapValues(list).collect()
```
