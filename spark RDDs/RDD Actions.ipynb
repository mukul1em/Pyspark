```
# actions 
# 1- Count()

from pyspark import SparkConf, SparkContext
conf = SparkConf().setAppName('count')
sc = SparkContext.getOrCreate(conf=conf)
```

```
rdd = sc.textFile('/FileStore/tables/filter.txt')
rdd.collect()
```

```
rdd.count() # there are 4 elements present in the rdd
```

```
rdd.flatMap(lambda x:x.split(" ")).count()
```

```
# 2 - countByValue 
#how many times each value occur in RDD
rdd.countByValue()
```

```
rdd.flatMap(lambda x : x.split(" ")).collect()
```

```
rdd.flatMap(lambda x : x.split(' ')).countByValue()
```
