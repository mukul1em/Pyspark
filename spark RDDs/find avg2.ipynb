```
from pyspark import SparkConf, SparkContext
conf = SparkConf().setAppName('average2')
sc = SparkContext.getOrCreate(conf=conf)
```

```
rdd = sc.textFile('/FileStore/tables/avg.csv')
rdd.collect()
```

```
def getkeyvalue(x):
    x = x.split(',')
    return (x[0], int(x[2][0]))
rdd2 = rdd.map(getkeyvalue)
```

```
rdd2 = rdd2.groupByKey().mapValues(list)
rdd2.collect()
```

```
def getavg(arr):
    month = arr[0]
    return (month, sum(arr[1])/len(arr[1]))
rdd2 = rdd2.map(getavg)
rdd2.collect()
```
