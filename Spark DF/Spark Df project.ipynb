```
from pyspark.sql.functions import min, max, avg,col
df.groupBy('department').agg(min(df.salary).alias('minium salary'), max(df.salary).alias('maximum salary')).orderBy(col('maximum salary').asc()).show()
```

```
df.filter(df.state=='NY').groupBy('state').agg(avg('bonus').alias('avg_bonus')).select('avg_bonus').collect()[0]['avg_bonus']
```

```
avg_bonus = df.filter(df.state=='NY').groupBy('state').agg(avg('bonus').alias('avg_bonus')).select('avg_bonus').collect()[0]['avg_bonus']
df.filter((df.state=='NY')&(df.department=='Finance')&(df.bonus > avg_bonus)).show()
```

```
#creating a udf
from pyspark.sql.functions import udf
from pyspark.sql.types import IntegerType
def increaseage(age, salary):
    if age > 45:
        return salary + 500
    else:
        return salary

inc_udf = udf(lambda x,y: increaseage(x,y), IntegerType())
df.withColumn('salary', inc_udf(df.age, df.salary)).show()
```
