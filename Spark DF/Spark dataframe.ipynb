```
from pyspark import SparkConf,SparkContext
conf = SparkConf().setAppName('RDd')
sc = SparkContext.getOrCreate(conf=conf)
rdd = sc.textFile('/FileStore/tables/studentdata.csv')
headers = rdd.first()
rdd = rdd.filter(lambda x: x!=headers).map(lambda x: x.split(','))
```

```
columns = headers.split(',')
dfrdd = rdd.toDF(columns)
dfrdd.printSchema()
```

```
from pyspark.sql.types import StructType, StructField, StringType, IntegerType
schema = StructType(
    [StructField('age',IntegerType(), True),
    StructField('gender', StringType(), True),
    StructField('name', StringType(), True),
    StructField('course', StringType(), True),
    StructField('roll', StringType(), True),
    StructField('marks', IntegerType(), True),
    StructField('email', StringType(), True)    
    ])

dfrdd = spark.createDataFrame(rdd, schema=schema)
dfrdd.printSchema()
```

```
df = spark.read.options(inferSchema='True', header='True').csv('/FileStore/tables/studentdata.csv')
df.select("gender","name").show()
```

```
df.select(df.name, df.email).show()
```

```
from pyspark.sql.functions import col
df.select(col('roll'),col('name')).show()
```

```
df.select('*').show()
```

```
#using the index
df.columns[2]
```

```
df.select(df.columns[0:2]).show()
```

```
df2 = df.withColumn('roll', col("roll").cast('string')) #return a new dataframe and old dataframe will remain as it is 
df2.printSchema()
```

```
#manipulate a column value
df = df.withColumn("marks", col("marks")+10) #marks are incremented by 10
df.show()
```

```
df = df.withColumn('marks', col('marks')-10)
df.show()
```

```
#creating a new column
df = df.withColumn('aggregated marks', col('marks'))
df.show()
```

```
#putting any hardcoded value inside the column
from pyspark.sql.functions import lit
df = df.withColumn("country", lit("USA"))
df.show()
```

```
#manipulating multiple columns 
df.withColumn("marks", col("marks")-10).withColumn('updated marks', col('marks') + 20).show()
```

```
df2 = df.withColumnRenamed('gender', 'sex')
df.show()
```

```
df.filter(df.course=='DB').show()
```

```
df.filter((df.course=='DB') & (df.marks > 50 )).show()
```

```
courses = ['DB', 'Cloud', 'OOP', 'DSA']
df.filter(df.course.isin(courses)).show()
```

```
#filter out the courses that are starting with D
df.filter(df.course.startswith('D')).show()
```

```
#filter out the courses that are endswith A
df.filter(df.name.endswith('e')).show()
```

```
#contains  a particular string 
df.filter(df.name.contains("se")).show()
```
