```
from pyspark.sql import SparkSession
from pyspark.sql.functions import col, lit
spark = SparkSession.builder.appName('Spark DataFrame').getOrCreate()
df = spark.read.option('header',True).csv('/FileStore/tables/studentdata.csv')
```

```
#creating a new column for total marks and let the total marks be 120
df2 = df.withColumn('total marks', lit(120))
#df2.show()
```

```
#
df2 = df2.withColumn('average marks', (col('marks')/col('total marks'))*100)
#df2.show()
```

```
#filter out those students who has achieved more than 80% marks in OOP course and save it in a new df
df3 = df.filter((df.course=='OOP')&(df.marks>80))
#df3.show()
```

```
df4 = df.filter((df.course=='Cloud')&(df.marks>60))
#df4.show()
```
