```
from pyspark.streaming import StreamingContext
from pyspark import SparkConf, SparkContext

conf = SparkConf().setAppName("spark streaming")
sc = SparkContext.getOrCreate(conf=conf)

ssc = StreamingContext(sc, 1) #1--> time duration it will look into the data directory and then take it on if new data comes up
```

```
rdd = ssc.textFileStream("/FileStore/tables/") #giving directory as input path
```

```
rdd.map(lambda x:(x,1))
rdd.reduceByKey(lambda x,y:x+y)
rdd.pprint() #rdd collect version of spark streaming 
ssc.start()
ssc.awaitTerminationOrTimeout(100)
```
